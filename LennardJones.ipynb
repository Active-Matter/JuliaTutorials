## Lennard-Jones Energy and Forces

Consider $N$ particles at positions $x_j \in \mathbb{R}^d$, then their collective energy under Lennard-Jones interaction is 
$$
\begin{align*}
  E(x) &= \sum_{i=1}^N \sum_{j \neq i} \phi(|x_i - x_j|), \\
  & \text{where} \quad \phi(r) = r^{-12} - 2 r^{-6}.
\end{align*}
$$
The vector of forces is simply the gradient $\nabla E(x)$, that is,
$$
  F_j := - \nabla_{x_j} E(x)
$$
is the force active on particle $x_j$.

```julia
# plot the LJ potential
using PyPlot
r = linspace(0.6, 3, 200)
subplot(2,2,1)
plot(r, r.^(-12) - 2 * r.^(-6), "b-")
plot(r, -12*r.^(-13)+12*r.^(-7), "r-")
axis([0, 3, -3, 3])
nothing
```

```julia
# Before we start testing, we set up a test problem
# we just make it an N x N 2D grid of particles.
function ljconfig(N, per=0.2)
    # mesh-grid-like construction
    X = (0:N) * ones(1,N+1)
    Y = X'
    # write all particles into a single array
    x = [X[:]', Y[:]']
    x = x + per * rand(size(x))
end

# plot the configuration
pygui(false)
N=50
x = ljconfig(N)
plot(squeeze(x[1,:],1), squeeze(x[2,:],1), "b.")
axis([-2, N+2, -2, N+2])
nothing
```

## Version 1: Plain C

We first test a plain C implementation. This will be the benchmark. It also shows off how easy it is to call C from Julia.

```julia
# first compile the C code - yes we can do this from Julia
run(`gcc lj_c.c -O3 -dynamiclib -o liblj_c`)
```

```julia
# The `wrapper function` lj_c just creates an empty array, passes it 
# to the C routine which fills it with the forces. The C routine
# also return E
function lj_c(x)
    dEc = zeros(size(x))
    Ec = ccall( (:energy, "./liblj_c"), 
                Cdouble, (Ptr{Cdouble}, Ptr{Cdouble}, Cint,      Cint), 
                      x,            dEc,          size(x,2), size(x,1))
    return Ec, dEc
end

# timing: run it once to compile it, then run it again to time it
Ec, dEc = lj_c(x)
@time Ec, dEc = lj_c(x)
nothing
```

## Version 2: A Convenient Modular Implementation

This first implementation is modular in that we call an external function that evaluates the LJ potential. Function calls in scripted languages tend to be *slow* (particularly so in Matlab!), so this is a useful test.

```julia
# Version 2: a convenient modular implementation
# This is the code I would like to write

# definition of the Lennard-Jones potential
function lj(r)
    s = sumabs2(r)
    J = s^(-6) - 2 * s^(-3)
    dJ = -12. * (s^(-7) - s^(-4)) * r
    return J, dJ
end

# definition of the total energy
function lj_pretty(x)
    E = 0.0
    dE = zeros(size(x))
    for n = 1:(size(x,2)-1), k = (n+1):size(x,2)
        J, dJ = lj(x[:,k]-x[:,n])
        E += J
        dE[:, k] += dJ
        dE[:, n] -= dJ
    end
    return E, dE
end


# And test it
E, dE = lj_pretty(x)
@time Ec, dEc = lj_pretty(x);
```

## Version 3: A C-"Style" Version

The "pretty" modular version is about a factor 100 slower than C. The profiler macro `@time` gives us a hint: somewhere we are allocating *a lot* of memoy. Julia's developers claim that we can speed it up by essentially writing C code (but in Julia). This is how it looks like.


```julia
function lj_cstyle(x)
    # the @inbounds macro turns off bound checking!
    @inbounds begin
    N = size(x,2); d = size(x,1)
    E = 0.0; dE = zeros(d, N)
    r = zeros(d);
    for n = 1:(N-1), k = (n+1):N
        s = 0.
        for i = 1:d
            r[i] = x[i,k]-x[i,n]
            s += r[i]*r[i]
        end
        t = 1./(s*s*s)
        E += t*t - 2.*t
        dJ = -12.*(t*t - t) / s
        for i = 1:d
            dE[i, k] += dJ * r[i]
            dE[i, n] -= dJ * r[i]
        end
    end
    end #inbounds
    return E, dE
end

# And test it
E, dE = lj_cstyle(x)
@time Ec, dEc = lj_cstyle(x);

# @profile lj_cstyle(x);
# using ProfileView
# ProfileView.view()
# Profile.print()
```

## Molecular Dynamics

We now solve the equations of motion
$$
  \ddot{x} = - \nabla E(x),
$$
where $E$ is the Lennard-Jones energy we implemented above. We use the simplest symplectic timestepping scheme (Euler A/B) for this.

Matlab-style "plot(); drawnow;" animations do work with Julia and PyPlot, but only in an external window and not within IJulia. The next cell shows how it is done.

```julia
# create a smaller configuration
using PyPlot
# plot in an external window since it doesn't work yet in IJulia
pygui(true)
# smallish system size
N=10
# initial configuration
x = ljconfig(N, 0.1)
p = zeros(size(x))
# time step
dt = 0.03
# total number of steps
nsteps = 100
# tell PyPlot that the plot is interactive
PyPlot.ion()
# . . . and that previous plots are overwritten
PyPlot.hold(false)
# start time-stepping loop
for n = 1:nsteps
    # plot current configuration
    plot(x[1,:], x[2,:], "k.", markersize=25)
    axis([-1, N+1, -1, N+1])
    title(string(" t = ", round(n*dt,2)))
    PyPlot.draw()

    # Euler-A method (or B; I forget)
    # forces
    E, dE = lj_cstyle(x)
    # velocity update
    p = p - dt * dE
    # position update
    x = x + dt * p
end
```

In order to show the animation within the IJulia notebook, we need to work a bit more. This builds on the [Interact](https://github.com/JuliaLang/Interact.jl) and Reactive packages. We first create a cell that is responsible for the plot, and then a second cell which does the compution and sends the output to the first cell. This kind of feels like a "hack" - I hope some alternatives will come along soon.

I use Winston instead of PyPlot since it is much faster, but one could equally use PyPlot as well, just change 
      `lift(myplot, myinput)`
to
      `lift(pyplot, myinput)`

```julia
using Reactive, Interact, Winston

# myfig = PyPlot.figure()
# function pyplot(data)
#     withfig(myfig) do
#         PyPlot.plot(data[1,:], data[2,:], "b.", markersize=25)
#         axis([-1,N+1,-1,N+1])
#     end
# end

function myplot(data)
    Winston.plot(data[1,:], data[2,:], "b.")
    Winston.xlim([-1,N+1])
    Winston.ylim([-1,N+1])
end

# smallish system size
N=30
# initial configuration
x = ljconfig(N, 0.1)
p = zeros(size(x))
# create an "input" object
myinput=Input(x)
lift(myplot, myinput)

# lift(pyplot, myinput)
```

```julia
# time step
dt = 0.03
# total number of steps
nsteps = 300
# start time-stepping loop
for n = 1:nsteps
    # plot current configuration
    push!(myinput, x)

    # Euler-A method (or B; I forget)
    # forces
    E, dE = lj_pretty(x)
    # velocity update
    p = p - dt * dE
    # position update
    x = x + dt * p
end
```
